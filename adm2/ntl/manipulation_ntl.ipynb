```python
from plotly.offline import init_notebook_mode
init_notebook_mode(connected=True)
```

```python
ntl = pd.read_csv("sum_ntl_adm2_2000_2021.csv")
```

```python
ntl_mean = pd.read_csv("mean_ntl_adm2_2000_2021.csv")
```

```python
fig = px.box(ntl, x = "shapeName", y = "total_sol")
fig.update_xaxes(categoryorder='total descending')
fig.show()
```

```python
ntl_wide = ntl_mean.pivot(index = ["id", "shapeName"], columns = "year", values = "total_sol")
```

```python
ntl_wide.head()
```

```python
ntl_2000 = ntl_wide[2000]
```

```python
ntl_2000.info()
```

```python
ntl_2016 = ntl_wide[2016]
```

```python
ntl_00and16 = pd.concat([ntl_2000, ntl_2016], axis = 1).add_prefix("ntl_")
```

```python
ntl_00and16.to_csv('2000and2016.csv', index=False)
```
